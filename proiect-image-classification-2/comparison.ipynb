```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10, 6))
plt.title("FashionMNIST")
plt.xlabel("Training Time (seconds)")
plt.ylabel("Test Accuracy")
plt.grid(True)
plt.xscale('linear')

plt.yticks([0.89, 0.89, 0.91, 0.92, 0.94, 0.87, 0.90, 0.87, 0.86])

points1 = [(200, 0.89, "MLP over features"), (170, 0.89, "MLP over images"), (450, 0.91, "CNN no augmentations"), (1102, 0.92, "CNN with augmentations"), (500, 0.94, "Fine-tuned ResNet"),]
points2 = [(11, 0.87, "Logistic Regression"), (290, 0.90, "SVM"), (290, 0.87, "Random Forest"), (265, 0.86, "Gradient Boosting")]
for x, y, label in points1:
    plt.scatter(x, y, label=label)

for x, y, label in points2:
    plt.scatter(x, y, label=label, marker='s')

plt.legend(loc="lower right")


plt.tight_layout()
plt.show()
```
